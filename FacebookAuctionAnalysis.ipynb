```python
import requests
import json
import pandas as pd

from secret import client_id, client_secret
from data_loader import DataLoader
```

```python
url = 'https://graph.facebook.com/oauth/access_token?grant_type=client_credentials&client_id={}&client_secret={}'\
    .format(client_id, client_secret)

r = requests.get(url)
access_token = r.text.split('=')
access_token = json.loads(access_token[0])['access_token']
print(access_token)
```

```python
album = {
    'boys': '1906437736340236',
    'girls': '1906437539673589',
    'guests': '1906438283006848',
    'seniors': '1906437936340216'
}
```

```python
data_loader = DataLoader(access_token)
```

```python
%%time
girls = data_loader.get_participants(album['girls'])
boys = data_loader.get_participants(album['boys'])
```

```python
print(len(girls), 'girls')
print(len(boys), 'boys')
```

```python
def create_table(participants):
    names = []
    descriptions = []
    last_comments = []

    for participant in participants:
        if len(participant.comments) > 1:
            last_comment_message = participant.comments[len(participant.comments) - 1].message
            names.append(participant.name)
            descriptions.append(participant.description)
            last_comments.append(last_comment_message)

    return pd.DataFrame(columns=['name', 'description', 'last_comment'],
                      data={'name': names, 'description': descriptions, 'last_comment': last_comments})
```

```python
df = create_table(girls)
df.to_csv('data/girls.csv')
df
```

```python
df = create_table(boys)
df.to_csv('data/boys.csv')
df
```
